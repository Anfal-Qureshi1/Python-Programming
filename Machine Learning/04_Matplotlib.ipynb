### Matplotlib workflow (object oriented approach)

```python
#1. import matplotlib
import matplotlib.pyplot as plt

#2. Prepare data
x = [1,2,3,4,5]
y = [2,4,6,8,10]

#3. Setup plot
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(x,y)

#4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

#5. Saving the plot
fig.savefig("savedFig")
```

### Making figures using numpy

```python
x= np.linspace(0,10,100)
x[:10]
```

```python
# Line plot
fig, ax = plt.subplots()
ax.plot(x, x**2)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
plt.show()
```

```python
# histogram
x = np.random.randn(5000)

fig, ax = plt.subplots()
ax.hist(x)
plt.show()
```

```python
# Subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 6))
ax1.plot(x)
ax2.scatter(x, x**2)
ax3.hist(x)
ax4.bar(x, x**2)
plt.show()
```

```python
# another option for making graph
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,6))

ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(17), np.random.random(17))
ax[1,0].bar(x, x**2)
ax[1,1].hist(np.random.randn(10000))
plt.show()
```

### When to use which type of plotting
* use pyplot for quick plots
* use object oriented method for complex plots (more flexible approach)

```python
 heart_disease
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50
```

```python
over_50.shape
```

```python
# pyplot method: plotting directly from the data frame
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target")
plt.show()
```

```python
# Object oriented method mixed with pyplot
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax)
ax.set_xlim([50,79])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
scatplot=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50['target'])

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
scatplot=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50['target'])

#Customization
ax.set(title="Heart Disease and Cholestrol Levels",
      xlabel="Age",
      ylabel="Cholestrol")
ax.legend(*scatplot.legend_elements(), title="Target") # returns handle and label

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
scatplot=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50['target'])

#Customization
ax.set(title="Heart Disease and Cholestrol Levels",
      xlabel="Age",
      ylabel="Cholestrol")
ax.legend(*scatplot.legend_elements(), title="Target")

# Adding horizontal line representing mean
ax.axhline(over_50["chol"].mean(), linestyle="--")

plt.show()
```

```python
over_50.head()
```

```python
# Subplot of cholestrol, age and thalach

fig, (ax0, ax1) = plt.subplots(nrows=2,
                                ncols=1,
                                figsize=(10, 10))
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

ax0.set(title="hear disease and cholestrol levels",
       xlabel="Age",
       ylabel="Cholestrol")

ax0.legend(*scatter.legend_elements(), title="target")
ax0.axhline(over_50["chol"].mean(),
    linestyle="--")


plt.show()
```

```python
# Subplot of cholestrol, age and thalach

fig, (ax0, ax1) = plt.subplots(nrows=2,
                                ncols=1,
                                figsize=(10, 10))
scatter1 = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

ax0.set(title="heart disease and cholestrol levels",
       xlabel="Age",
       ylabel="Cholestrol")

ax0.legend(*scatter1.legend_elements(), title="target")
ax0.axhline(over_50["chol"].mean(),
    linestyle="--")

scatter2 = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])
ax1.set(title="heart disease and thalach levels",
       xlabel="Age",
       ylabel="Max Heart Rate")
ax1.legend(*scatter2.legend_elements(), title="target")
ax1.axhline(over_50["thalach"].mean(),
           linestyle="--")

fig.suptitle("Heart Disease Data Analaysis", fontsize=12, fontweight="bold")
plt.show()
```
