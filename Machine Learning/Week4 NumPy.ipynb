## Normalization in Numpy

```python
X = ecg[:, :187]   # features
y = ecg[:, 187]    # target
```

```python
mean_per_row = np.mean(X, axis=1, keepdims=True)
std_per_row  = np.std(X, axis=1, keepdims=True)
```

```python
mean_per_row.shape, std_per_row.shape
```

```python
X_norm = (X - mean_per_row) / std_per_row
```

```python
X_norm.shape
```

```python
np.mean(X_norm, axis=1)
```

```python
np.std(X_norm, axis=1)
```
