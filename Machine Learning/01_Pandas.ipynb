### comparing each category of waveform

## Standardization/Normalization of data

```python
ecg_col_mean = X.mean()
ecg_col_std = X.std()
```

```python
X_norm = (X - ecg_col_mean) / ecg_col_std

X_norm.describe().loc[['mean','std']].head()
```

```python
X_norm = (X - ecg_col_mean) / ecg_col_std

X_norm.describe().loc[['mean','std']].head()
```

```python
X_norm.head()
```

### Reassign X (standardized), and y

```python
X = X_norm
y = ecg.iloc[:, -1]
```

```python
X.head()
```

```python
y.value_counts()
```

### Exploring shape of waveform - classwise

```python
class_means = X.groupby(y).mean()

class_means.T.plot(figsize=(12,6), title="Average ECG per Class")
```

```python
class_means = X_norm.groupby(y).mean()

class_means.T.plot(figsize=(12,6), title="Average ECG per Class")
```
